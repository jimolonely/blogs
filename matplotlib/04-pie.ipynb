```python
import matplotlib.pyplot as plt
```

```python
labels = ['A','B','C','D']
sizes = [10,20,30,40]

```

```python
fig,ax = plt.subplots()
ax.pie(sizes,labels=labels)

plt.show()
```

```python
fig,ax = plt.subplots()
# 着重突出
explode = (0,0.1,0,0)
ax.pie(sizes,explode=explode,labels=labels)

plt.show()
```

```python
# 显示阴影并显示百分比
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True)

plt.show()
```

```python
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True)

ax.axis('equal') # 保证是个圆

plt.show()
```

```python
# 调整角度
fig,ax = plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)

ax.axis('equal') # 保证是个圆

plt.show()
```
