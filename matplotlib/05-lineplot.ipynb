```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 准备数据和函数
x = np.arange(0.0,2.0,0.01)
y = 1 + np.sin(2*np.pi*x)

#绘制
fig,ax = plt.subplots() # 等价于fig = plt.figure ; ax = fig.add_subplot(111)
ax.plot(x,y)

# 设置属性
ax.set(xlabel='time(s)',ylabel='voltage(mV)',title='line plot')
ax.grid()

fig.savefig('01-line.png')
plt.show()
```
