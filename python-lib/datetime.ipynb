# datetime日期处理

```python
import datetime
help(datetime)
```

# 1.设定日期

```python
max = datetime.date(9999,12,31)
min = datetime.date(1,1,1)
print(max,min)
print(type(max))
```

```python
re = datetime.timedelta(1)
print(re)
```

# 2.设定日期和时间

```python
max = datetime.datetime(9999,12,31,23,59,59,999999)
min = datetime.datetime(1,1,1,0,0)
print(max,min)
print(type(max))
```

# 3.设定时间

```python
max = datetime.time(23,59,59,999999)
min = datetime.time(0,0)
print(max,min)
print(type(max))
```

# 4.时间偏移

```python
d = datetime.datetime.now()
print(d)
max = datetime.timedelta(1)
```

# 5.时区

```python
z = datetime.timezone(datetime.timedelta())
print(z)
z = datetime.timezone(datetime.timedelta(0,86340))
print(z)
z = datetime.timezone(datetime.timedelta(-1,60))
print(z)
print(datetime.timezone.utc)
```
