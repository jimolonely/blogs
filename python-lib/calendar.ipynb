# 提供一组和日期相关的函数

```python
import calendar
help(calendar)
```

### 获得某年日历

```python
cal = calendar.calendar(2017)
print(cal)
```

### 获得某月日历

```python
cal = calendar.month(2017,10)
print(cal)
```

### 返回矩阵

```python
print(calendar.monthcalendar(2017,10))
```

### 是否是闰年

```python
print(calendar.isleap(2017))
```

### 有多少个闰年

```python
print(calendar.leapdays(1900,2018))
```
